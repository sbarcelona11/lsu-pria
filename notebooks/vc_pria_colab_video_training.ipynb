# VC-PRIA — Entrenamiento desde videos en Google Colab

Este notebook permite entrenar el proyecto directamente desde videos etiquetados guardados en Google Drive.

Layouts soportados:
- `label`: `root/<label>/clip.mp4`
- `subject_label`: `root/<subject>/<label>/clip.mp4`


```python
REPO_URL = "https://github.com/REEMPLAZAR/VC-pria.git"
WORK_MODE = "frame"  # "frame" | "multimodal"
VIDEOS_ROOT = "/content/drive/MyDrive/VC-pria/data/train_videos"
WORK_DIR = "/content/drive/MyDrive/VC-pria/runs/colab_video_training"
LAYOUT = "label"  # "label" | "subject_label" | "auto"
USE_PREPROCESS = True
USE_SKIN_MASK = True
CNN_DEVICE = "cuda"
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os, torch
print('cuda available:', torch.cuda.is_available())
print('videos root exists:', os.path.exists(VIDEOS_ROOT))
```

```python
cmd = [
    'python3', 'vcpria.py', 'train-videos-stack',
    '--videos-root', VIDEOS_ROOT,
    '--work-dir', WORK_DIR,
    '--layout', LAYOUT,
    '--cnn-device', CNN_DEVICE,
]
if USE_PREPROCESS:
    cmd.append('--preprocess')
if USE_SKIN_MASK:
    cmd.append('--skin-mask')
print(' '.join(cmd))
```

```python
mm_cmd = [
    'python3', 'vcpria.py', 'train-videos-mm-stack',
    '--videos-root', VIDEOS_ROOT,
    '--work-dir', WORK_DIR,
    '--layout', LAYOUT,
]
if USE_PREPROCESS:
    mm_cmd.append('--preprocess')
print(' '.join(mm_cmd))
```

```python
import subprocess
active = cmd if WORK_MODE == 'frame' else mm_cmd
subprocess.run(active, check=True)
```

```python
import os
for root, _, files in os.walk(WORK_DIR):
    for name in files:
        if name.endswith(('.joblib', '.pt', '.json', '.md', '.png')):
            print(os.path.join(root, name))
```
